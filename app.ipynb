```python
from flask import Flask, request, render_template
import pickle

app = Flask(__name__)
model = pickle.load(open('model.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('index.html')


@app.route('/predict',methods=['POST'])
def predict():
    
    Age = int(request.form["Age"])
    Height = int(request.form["Height"])
    Weight = int(request.form["Weight"])
    AerialAbility = int(request.form["AerialAbility"])
    CommandOfArea = int(request.form["CommandOfArea"])
    Communication = int(request.form["Communication"])
    Eccentricity = int(request.form["Eccentricity"])
    Handling = int(request.form["Handling"])
    Kicking = int(request.form["Kicking"])
    OneOnOnes = int(request.form["OneOnOnes"])
    Reflexes = int(request.form["Reflexes"])
    RushingOut = int(request.form["RushingOut"])
    TendencyToPunch = int(request.form["TendencyToPunch"])
    Throwing = int(request.form["Throwing"])
    Corners = int(request.form["Corners"])
    Crossing = int(request.form["Crossing"])
    Dribbling = int(request.form["Dribbling"])
    Finishing = int(request.form["Finishing"])
    FirstTouch = int(request.form["FirstTouch"])
    Freekicks = int(request.form["Freekicks"])
    Heading = int(request.form["Heading"])
    LongShots = int(request.form["LongShots"])
    Longthrows = int(request.form["Longthrows"])
    Marking = int(request.form["Marking"])
    Passing = int(request.form["Passing"])
    PenaltyTaking = int(request.form["PenaltyTaking"])
    Tackling = int(request.form["Tackling"])
    Technique = int(request.form["Technique"])
    Aggression = int(request.form["Aggression"])
    Anticipation = int(request.form["Anticipation"])
    Bravery = int(request.form["Bravery"])
    Composure = int(request.form["Composure"])
    Concentration = int(request.form["Concentration"])
    Vision = int(request.form["Vision"])
    Decisions = int(request.form["Decisions"])
    Determination = int(request.form["Determination"])
    Flair = int(request.form["Flair"])
    Leadership = int(request.form["Leadership"])
    OffTheBall = int(request.form["OffTheBall"])
    Positioning = int(request.form["Positioning"])
    Teamwork = int(request.form["Teamwork"])
    Workrate = int(request.form["Workrate"])
    Acceleration = int(request.form["Acceleration"])
    Agility = int(request.form["Agility"])
    Balance = int(request.form["Balance"])
    Jumping = int(request.form["Jumping"])
    LeftFoot = int(request.form["LeftFoot"])
    NaturalFitness = int(request.form["NaturalFitness"])
    Pace = int(request.form["Pace"])
    RightFoot = int(request.form["RightFoot"])
    Stamina = int(request.form["Stamina"])
    Strength = int(request.form["Strength"])
    Consistency = int(request.form["Consistency"])
    Dirtiness = int(request.form["Dirtiness"])
    ImportantMatches = int(request.form["ImportantMatches"])
    InjuryProness = int(request.form["InjuryProness"])
    Versatility = int(request.form["Versatility"])
    Adaptability = int(request.form["Adaptability"])
    Ambition = int(request.form["Ambition"])
    Loyalty = int(request.form["Loyalty"])
    Pressure = int(request.form["Pressure"])
    Professional = int(request.form["Professional"])
    Sportsmanship = int(request.form["Sportsmanship"])
    Temperament = int(request.form["Temperament"])
    Controversy = int(request.form["Controversy"])
    
    position = model.predict([[Age, Height, Weight, AerialAbility, CommandOfArea, Communication, Eccentricity, 
                               Handling, Kicking, OneOnOnes, Reflexes, RushingOut, TendencyToPunch, Throwing, Corners, Crossing,
                                Dribbling, Finishing, FirstTouch, Freekicks, Heading, LongShots, Longthrows, Marking, Passing, PenaltyTaking,
                             Tackling, Technique, Aggression, Anticipation, Bravery, Composure, Concentration, Vision, Decisions, Determination,
                              Flair, Leadership, OffTheBall, Positioning, Teamwork, Workrate, Acceleration, Agility, Balance, 
                              Jumping, LeftFoot, NaturalFitness, Pace, RightFoot, Stamina, Strength, Consistency, Dirtiness, 
                               ImportantMatches, InjuryProness, Versatility, Adaptability, Ambition, Loyalty, Pressure, Professional,
                               Sportsmanship, Temperament, Controversy]])  

    
    probabilities = model.predict_proba([[Age, Height, Weight, AerialAbility, CommandOfArea, Communication, Eccentricity, 
                               Handling, Kicking, OneOnOnes, Reflexes, RushingOut, TendencyToPunch, Throwing, Corners, Crossing,
                                Dribbling, Finishing, FirstTouch, Freekicks, Heading, LongShots, Longthrows, Marking, Passing, PenaltyTaking,
                             Tackling, Technique, Aggression, Anticipation, Bravery, Composure, Concentration, Vision, Decisions, Determination,
                              Flair, Leadership, OffTheBall, Positioning, Teamwork, Workrate, Acceleration, Agility, Balance, 
                              Jumping, LeftFoot, NaturalFitness, Pace, RightFoot, Stamina, Strength, Consistency, Dirtiness, 
                               ImportantMatches, InjuryProness, Versatility, Adaptability, Ambition, Loyalty, Pressure, Professional,
                               Sportsmanship, Temperament, Controversy]]).tolist()[0]  

    probabilities_pct = [round(element * 100, 1) for element in probabilities]
    defender = probabilities_pct[0]
    goalkeaper = probabilities_pct[1]
    midfielder = probabilities_pct[2]
    striker = probabilities_pct[3]
    
    return render_template('index.html', prediction_text=f'Your optimal position in soccer is: {position[0]}. You are {goalkeaper}% GK, {defender}% DF, {midfielder}% M, and {striker}% ST.')

if __name__ == "__main__":
    app.run()
```
